```python
# Mann-Whitney U test

from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(df[df['height']>69]['height'], df[df['height']<=69]['height'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distribution (fail to reject H0)')
else:
	print('Different distribution (reject H0)')
```

```python
# Kruskal-Wallis H-test
from scipy.stats import kruskal


# compare samples
stat, p = kruskal(df[df['height']>69]['height'], df[df['height']<=69]['height'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

# poly

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df.columns
```

```python
y = df['IOCT_Time'].values
X = df[['height','weight' ]].values
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X,y)
lin_reg.coef_
```

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
lin_reg2.coef_
```

```python
y_poly_pred = lin_reg2.predict(X_poly)
r2_score(y, y_poly_pred)
```

```python
np.sqrt(mean_squared_error(y, y_poly_pred))
```

```python
for i in range(2,20):

  poly_reg = PolynomialFeatures(degree=i)
  X_poly = poly_reg.fit_transform(X)
  lin_regN = LinearRegression()
  lin_regN.fit(X_poly,y)
  print(i,r2_score(y, lin_regN.predict(X_poly)), np.sqrt(mean_squared_error(y, lin_regN.predict(X_poly))))
```
