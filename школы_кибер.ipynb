```python
# Mann-Whitney U test

from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(df_neuch.iloc[:, 0:1], df_neuch_kiber.iloc[:, 0:1])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distribution (fail to reject H0)')
else:
	print('Different distribution (reject H0)')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#create a copy of the dataframe
df_copy = df.drop([ 'Дисциплина', 'Дата.заполнения'], axis=1)

#drop the index column from the dataframe
#ddf_copy.drop('index', axis=1, inplace=True)

#fit the KMeans model to the dataframe
kmeans = KMeans(n_clusters=3).fit(df_copy)

#add the cluster labels to the dataframe
df_copy['cluster'] = kmeans.labels_

#calculate the silhouette score for the model
from sklearn.metrics import silhouette_score
silhouette_score(df_copy, kmeans.labels_)

#visualize the clusters using a scatter plot
plt.scatter(df_copy['Кол.школ'], df_copy['Кол.команд'], c=kmeans.labels_, cmap='rainbow')
plt.xlabel('Number of Schools')
plt.ylabel('Number of Teams')
plt.title('Clusters of Schools and Teams')
plt.show()
silhouette_score(df_copy, kmeans.labels_)
```
