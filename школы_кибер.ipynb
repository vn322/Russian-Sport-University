```python
df_neuch_kiber = df.groupby(['7.7']).agg('count') #В чем вы видите причины неучастия образовательной организации в мероприятиях
df_neuch_kiber.sort_values(['6.2'],ascending=False)
df_neuch_kiber.iloc[:, 0:1].sort_values(['Сокращённое наименование организации'],ascending=False)
```

```python
# Mann-Whitney U test

from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(df_neuch.iloc[:, 0:1], df_neuch_kiber.iloc[:, 0:1])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distribution (fail to reject H0)')
else:
	print('Different distribution (reject H0)')
```
